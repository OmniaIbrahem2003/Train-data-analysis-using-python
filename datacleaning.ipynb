```python
import pandas as pd
import re
```

```python
df = pd.read_csv('Data_Train.csv')
```

```python
df.isna().sum().sum()
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
df.dropna(inplace=True,ignore_index=True)
```

```python
df['Total_Stops'] = df['Total_Stops'].replace('non-stop', '0')
df['Total_Stops'] = df['Total_Stops'].replace(r'(\d+)\s*stops?', r'\1', regex=True)
```

```python
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
```

```python
df["Duration"][:10]
```

```python
def convert_duration_to_minutes(duration):
    match = re.match(r'(?:(\d+)h)?\s*(?:(\d+)m)?', duration)
    hours = int(match.group(1)) if match.group(1) else 0 
    minutes = int(match.group(2)) if match.group(2) else 0 
    total_minutes = hours * 60 + minutes
    return total_minutes
```

```python
df['Duration'] = df['Duration'].apply(convert_duration_to_minutes)
```
